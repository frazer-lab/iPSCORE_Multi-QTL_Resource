```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib_venn import venn3
```

```python
atac = pd.read_csv("/frazer01/home/tarthur/projects/Resource/SupplementalTables/TableS3_ATACMetadata_v2.txt", sep = "\t")
rna  = pd.read_csv("/frazer01/home/tarthur/projects/Resource/SupplementalTables/TableS2_RNAMetadata_v2.txt", sep = "\t")
chip = pd.read_csv("/frazer01/home/tarthur/projects/Resource/SupplementalTables/TableS4_ChIPMetadata_v2.txt", sep = "\t")
```

```python
atac.columns = atac.columns.str.replace(" ", "_")
rna.columns = rna.columns.str.replace(" ", "_")
chip.columns = chip.columns.str.replace(" ", "_")
```

```python
chip.head(2)
```

```python
t = 'iPSC'

set1 = set(atac[atac.Tissue == t].iPSC_line_ID.tolist())
set2 = set(rna[rna.Tissue == t].iPSC_Line_ID.tolist())
set3 = set(chip[chip.Tissue == t].iPSC_line_ID.tolist())

plt.figure(figsize = (6,6))
v = venn3_unweighted([set1, set2, set3], ["ATAC", "RNA", "H3K27ac"])

for label in v.subset_labels + v.set_labels:
    if label:  # Checking if label is not None
        label.set_fontsize(15)
        label.set_family("arial")

plt.title("Overlap of iPSCs across Data Types", fontsize = 16)
plt.show()
```
