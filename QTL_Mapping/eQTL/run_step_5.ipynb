# Check

```r
# check that conds are sequential
tmp = qtls %>% filter(egene == T) 
a = data.frame(rbindlist(lapply(tmp$element_id, function(x) 
{
    this = tmp %>% filter(element_id == x)
    types = this$type
    seqs = c(min(types) : max(types))
    data.frame(element_id = x, nmissing = length(seqs[which(!seqs %in% types)] ))
}))) %>% arrange(desc(nmissing))
table(a$nmissing)
```

```r
# check lead duplicates
tmp %>% mutate(id = paste(element_id, id, sep = "_")) %>% filter(duplicated(id))
```
